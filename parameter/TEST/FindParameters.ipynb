```python
import pandas as pd
path = 'DataForThai.csv'
df = pd.read_csv(path)
```

```python
df
```

```python
df_thai = df.copy()
```

```python
import numpy as np
recovered = [i for i in df_thai.Recovered]
recovered = np.array(recovered, dtype=float)
recovered
```

```python
deaths = [i for i in df_thai.Deaths]
deaths = np.array(deaths, dtype=float)
deaths
```

```python
confirmed = [i for i in df_thai.Confirmed]
confirmed = np.array(confirmed, dtype=float)
confirmed
```

```python
data = confirmed - deaths - recovered
data
```

```python
# print(data)
# data = confirmed

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from lmfit import minimize,  Parameters, Parameter, report_fit

def f(y, t , paras):
    
    S  = y[0]
    I = y[1]
    R = y[2]
    
    try:
        beta = paras['beta'].value
        gamma = paras['gamma'].value
    except KeyError:
        beta, gamma = paras
        
    # the model equations
    ds_dt = - beta*I*S
    di_dt = beta*I*S - gamma*I
    dr_dt = I*gamma
    return [ds_dt, di_dt, dr_dt]

def g(t, x0, paras):
    x = odeint(f, x0, t, args=(paras,))
    return x
   
def residual(paras, t, data):
    x0 = paras['S0'].value, paras['I0'].value, paras['R0'].value
    model = g(t, x0, paras)
    
    x2_model = model[:,1]
#     print('x2_model',x2_model.shape)
#     print('data',data.shape)
#     print('x2_model = ',x2_model)
#     print('data = ',data)
    return (x2_model - data).ravel()

# initial conditions
# S0 = np.max(data)
S0 = 5000.
I0 = 2.
R0 = 0.0
y0 = [S0, I0, R0]

# measured data
t_measured = np.linspace(0, len(data)-1, len(data))
x2_measred = data

plt.figure()
plt.scatter(t_measured, x2_measred, marker='o', color='b', label='measured data', s=25)

# set parameters including bounds; you can fit parameter (use very = False)
params = Parameters()
params.add('S0', value=S0, vary=False)
params.add('I0', value=I0, vary = False)
params.add('R0', value=R0, vary = False)
params.add('beta', value=0.001, min=0.0000000000001, max=2.)
params.add('gamma', value=.0001, min=0.000000001, max=2.)

# fit model
# result = minimize(residual, params, args=(t_measured, x2_measred), method='leastsq')
result = minimize(residual, params, args=(t_measured, x2_measred), method= 'Nelder-Mead')

# data_fitted = g(np.linspace(0.,9.,1000), y0, result.params)
data_fitted = g(np.arange(len(data)+1), y0, result.params)
# print('data_fitted = ',data_fitted)
# plot fitted data
# plt.plot(np.linspace(0., 9., 1000), data_fitted[:,1], '-', linewidth=2, color='red', label='fitted data')
plt.plot(np.arange(len(data)+1), data_fitted[:,1], '-', linewidth=2, color='red', label='fitted data')
plt.legend()
# plt.xlim([0, max(t_measured)])
# plt.ylim([0, 1.1*max(data_fitted[:,1])])

report_fit(result)

plt.show
```
